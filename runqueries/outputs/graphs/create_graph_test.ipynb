```python
# Importar bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leer datos desde el archivo CSV
data = pd.read_csv('../ranked/louds/backtracking/results.csv', delimiter=';')
data
```

```python
# Variables del archivo
variables = data.columns[1:]  # Excluir la columna 'k'
variables
```

```python
for variable in variables:
    print("variable:", variable)
    print(data[variable])
    print("\n")
    
    
```

```python
# Crear el gráfico
plt.figure(figsize=(14, 8))

for variable in variables:
    plt.plot(data['k'], data[variable], marker='o', label=variable)
        
# Configuración del gráfico
plt.xscale('log')  # Escala logarítmica en el eje x
plt.xlabel('Valor de k')
plt.ylabel('Tiempo (ms)')
plt.title('Curvas de tiempo en función del valor de k')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
